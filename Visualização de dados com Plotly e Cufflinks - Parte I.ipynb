## Visualização de dados com Plot.ly

```python
#Importação das bibiotecas Cufflinks, Plot.ly, Pandas e Numpy

import cufflinks as cf
from plotly.offline import plot, iplot
import pandas as pd
import numpy as np
```

```python
#Criação de um dataframe com 100 números aleatórios divididos entre as colunas A, B, C e D

df = pd.DataFrame(np.random.randn(100,4), columns=['A','B','C','D'])
```

```python
#Visualização das 5 primeiras linhas do dataframe

df.head()
```

```python
#Habilitação do modo offline do Cufflink

cf.go_offline()
```

```python
#Visualização dos dados offline

df.iplot()
```

**Gráficos do tipo scatter**

```python
df.iplot(kind='scatter', x='A',y='B', mode='markers')
```

**Gráficos de Barras**

```python
df.sum().iplot(kind='bar')
```

**Gráficos de Boxplots**

```python
df.iplot(kind='box')
```

**Histogramas**

```python
df['A'].iplot(kind='hist')
```

**Histograma aninhado**

```python
df.iplot(kind='hist')
```

**Spread plots**

```python
df[['A','D','C','B']].iplot(kind='spread')
```

**Bubble Plots**

```python
df.iplot(kind='bubble', x='A',y='B', size='C', colors='seagreen')
```

**Scatter Matrix**

```python
df.scatter_matrix()
```
